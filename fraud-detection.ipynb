# Feature Engineering : # Creating New Feature - 1. Transaction Time Feature

```python
df['hour'] = df['step'] % 24
df['is_night'] = df['hour'].apply(lambda x : 1 if x < 6 else 0)
```

```python
df.head()
```

# 2. Transaction Amount Feature

```python
df['amount_ratio'] = df['amount']/(df['oldbalanceOrg'] + 1)
```

```python
df.head()
```

# 3. Balance Difference Feature

```python
df['sender_balance_change'] = df['oldbalanceOrg'] - df['newbalanceOrig']
df['receiver_balance_change'] = df['newbalanceDest'] - df['oldbalanceDest']
```

```python
df[df['isFraud']==1]
```

# 4. Unexpected Zero Balance

```python
df['orig_balance_zero'] = (df['oldbalanceOrg'] == 0).astype(int)
df['dest_balance_zero'] = (df['oldbalanceDest'] == 0).astype(int)
```

# 2. Encoding Categorical Feature

```python
dummies = pd.get_dummies(df['type']).astype(int)
dummies.tail()
```

```python
df = pd.get_dummies(df, columns=['type'], drop_first=True)
```

```python
df['type_TRANSFER'] = df['type_TRANSFER'].astype(int)
df.head()
```

# Independent and dependent feature

```python
df.drop(columns=['step'], inplace=True)
```

```python
x = df.drop('isFraud', axis=1)
y = df['isFraud']
```

```python
x
```

# Train - Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state= 42, shuffle=True)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Model Training

# Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

import joblib
joblib.dump(scaler, "scaler_latest.pkl")
```

```python
x_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
```

```python
lr.fit(x_train_scaled, y_train)
```

```python
y_pred1 = lr.predict(x_test_scaled)
y_pred1
```

# Check the classification report of the logistic regression model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

acc_score = accuracy_score(y_test, y_pred1)
print(f"accuracy score : {acc_score} ")
print()
print("Confusion matrix :")
print(confusion_matrix(y_test, y_pred1))
print()
print("Classification Report :")
print(classification_report(y_test, y_pred1))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import joblib

pipe = Pipeline([
    ("Scaler", StandardScaler()),
    ("model", RandomForestClassifier(n_estimators = 100, class_weight={0: 1, 1: 10}, random_state =42))
])

pipe.fit(x_train, y_train)

#Evaluate
pred_test = pipe.predict(x_test)
print(classification_report(y_test, pred_test))

# save model
joblib.dump(pipe, "fraud_pipeline.pkl")
print("Pipeline saved successfully!")
```

```python
import joblib

pipe_loaded = joblib.load("fraud_pipeline.pkl")
y_pred_loaded = pipe_loaded.predict(x_test)
print(classification_report(y_test, y_pred_loaded))
```
